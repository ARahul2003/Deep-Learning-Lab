**Define functions to add or subtract the 5D tensors made using videos**

```python
def subtractVideos(x1, x2):
  assert x1.ndim == 5
  assert x2.ndim == 5
  for i1 in range(x1.shape[0]):
    for i2 in range(x1.shape[1]):
      for i3 in range(x1.shape[2]):
        for i4 in range(x1.shape[3]):
          for i5 in range(x1.shape[4]):
            if x1[i1][i2][i3][i4][i5] > x2[i1][i2][i3][i4][i5]:
              x1[i1][i2][i3][i4][i5] -= x2[i1][i2][i3][i4][i5]
  
  return x1
```

```python
def addVideos(x1, x2):
  assert x1.ndim == 5
  assert x2.ndim == 5
  for i1 in range(x1.shape[0]):
    for i2 in range(x1.shape[1]):
      for i3 in range(x1.shape[2]):
        for i4 in range(x1.shape[3]):
          for i5 in range(x1.shape[4]):
            if x1[i1][i2][i3][i4][i5] + x2[i1][i2][i3][i4][i5] < 255:
              x1[i1][i2][i3][i4][i5] += x2[i1][i2][i3][i4][i5]
  
  return x1
```

# **Perceptron network for simple adder circuit**

```python
def thres(x):
  if x > 0:
    return 1
  else:
    return 0

def perceptronModel(x, w, b):
  y = np.dot(w,x)
  y = y + b
  y = thres(y)
  return y

def nand(x):
  w = np.array([-3,-2])
  b = 4
  return perceptronModel(x,w,b)

def xor(x):
  return nand([nand([x[0], nand(x)]), nand([nand(x),x[1]])])

def And(x):
  return nand([nand(x), nand(x)])

def adder(x):
  return [xor(x), And(x)]  
```

```python
print("Addition of 0 and 0 gives: ", adder([0,0])[0], ", carry is:", adder([0, 0])[1])
print("Addition of 0 and 1 gives: ", adder([0,1])[0], ", carry is:", adder([0, 1])[1])
print("Addition of 1 and 0 gives: ", adder([1,0])[0], ", carry is:", adder([1, 0])[1])
print("Addition of 1 and 1 gives: ", adder([1,1])[0], ", carry is:", adder([1, 1])[1])
```
